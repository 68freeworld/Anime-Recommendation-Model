```python
print("\n### Recommendations Based on Popularity (Most Rated) ###")

# Count how many times each anime appears in the ratings dataframe
anime_popularity = rating_df_clean['anime_name'].value_counts()

# Get the top 10 most frequently rated anime
top_10_popular_anime = anime_popularity.head(10)

print("Top 10 Most Popular Anime (by number of ratings):")
print(top_10_popular_anime)
print("-" * 30)
```

```python
print("\n### Recommendations Based on Average Rating (Simple) ###")

# Calculate the average rating for each anime
# Group by anime_name and calculate the mean of the 'rating' column
avg_rating_df = rating_df_clean.groupby('anime_name')['rating'].mean()

# Sort the results by average rating in descending order
sorted_avg_rating_df = avg_rating_df.sort_values(ascending=False)

print("Top 10 Highest Rated Anime (Simple Average):")
print(sorted_avg_rating_df.head(10))
print("-" * 30)
```

```python
print("Checking number of ratings for the top simple average rated anime:")
top_avg_rated_names = sorted_avg_rating_df.head(10).index
ratings_count_for_top_avg = rating_df_clean[rating_df_clean['anime_name'].isin(top_avg_rated_names)]['anime_name'].value_counts()
print(ratings_count_for_top_avg)
print("(Notice: Some might have very few ratings, making the high average less reliable)")
print("-" * 30)
```

```python
print("\n### Recommendations Based Popularity-Driven Average Rating ###")

#  calculate average ratings and counts for all anime
anime_stats = rating_df_clean.groupby('anime_name').agg(
    Avg_Rating=('rating', 'mean'),
    Num_Ratings=('rating', 'count')
).reset_index()


# Calculate the mean or median number of ratings to get a sense of distribution
mean_ratings = anime_stats['Num_Ratings'].mean()
median_ratings = anime_stats['Num_Ratings'].median()
print(f"Mean number of ratings per anime: {mean_ratings:.2f}")
print(f"Median number of ratings per anime: {median_ratings:.2f}")

threshold = 200
print(f"Using a minimum rating threshold of: {threshold}")

# Filter the stats DataFrame to include only anime with enough ratings
frequent_anime_stats = anime_stats[anime_stats['Num_Ratings'] >= threshold]

# Sort these frequently rated anime by their average rating
frequent_anime_sorted = frequent_anime_stats.sort_values(by='Avg_Rating', ascending=False)

print(f"\nTop 10 Highest Rated Anime (with at least {threshold} ratings):")
print(frequent_anime_sorted[['anime_name', 'Avg_Rating', 'Num_Ratings']].head(10))
print("-" * 30)
```

```python
print("\n### Recommendations Based on Weighted Scoring ###")



max_num_ratings = anime_stats['Num_Ratings'].max()
max_avg_rating = anime_stats['Avg_Rating'].max()

if max_num_ratings == 0 or max_avg_rating == 0:
    print("Warning: Max number of ratings or max average rating is zero. Cannot calculate weighted score.")
else:
    # Normalize the number of ratings and average ratings
    anime_stats['Norm_Num_Ratings'] = anime_stats['Num_Ratings'] / max_num_ratings
    anime_stats['Norm_Avg_Rating'] = anime_stats['Avg_Rating'] / max_avg_rating # Using max observed avg rating

    # Define weights for popularity vs. rating
    W1 = 0.5  # Popularity weight
    W2 = 0.5  # Average rating weight
    print(f"Using Weights: Popularity (W1) = {W1}, Average Rating (W2) = {W2}")

    # Calculate the weighted score
    anime_stats['Weighted_Score'] = (W1 * anime_stats['Norm_Num_Ratings']) + (W2 * anime_stats['Norm_Avg_Rating'])

    # Sort by the weighted score
    weighted_ranked_anime = anime_stats.sort_values(by='Weighted_Score', ascending=False)

    print("\nTop 10 Anime based on Weighted Score:")
    print(weighted_ranked_anime[['anime_name', 'Avg_Rating', 'Num_Ratings', 'Weighted_Score']].head(10))
    print("-" * 30)
```

```python
print("\n### Recommendations Based on Bayesian Scoring ###")



# Calculate global average rating (m)
global_avg_rating = rating_df_clean['rating'].mean()
print(f"Global average rating (m): {global_avg_rating:.4f}")

# Calculate the mean number of ratings across all anime (C - confidence factor)
# Using the previously calculated anime_stats['Num_Ratings']
C = anime_stats['Num_Ratings'].mean()
print(f"Mean number of ratings per anime (C): {C:.4f}")

# Apply the Bayesian Rating formula
# Bayesian Rating = (C * m + Num_Ratings * Avg_Rating) / (C + Num_Ratings)
anime_stats['Bayesian_Rating'] = (
    (C * global_avg_rating + anime_stats['Num_Ratings'] * anime_stats['Avg_Rating']) /
    (C + anime_stats['Num_Ratings'])
)

# Sort by the Bayesian rating
bayesian_ranked_anime = anime_stats.sort_values(by='Bayesian_Rating', ascending=False)

print("\nTop 10 Anime based on Bayesian Rating:")
print(bayesian_ranked_anime[['anime_name', 'Avg_Rating', 'Num_Ratings', 'Bayesian_Rating']].head(10))
print("-" * 30)

print("\nNon-personalized recommendation generation complete.")
```

```python
# =============================================================
# 1. HELPER FUNCTIONS
# =============================================================
def add_weighted_score(df, w1=0.5, w2=0.5):
    """
    Adds column 'Weighted_Score' to *df* using your W1/W2 formula.
    """
    max_n = df["Num_Ratings"].max()
    max_r = df["Avg_Rating"].max()

    if max_n == 0 or max_r == 0:
        raise ValueError("Max ratings or max score is zero â€“ cannot normalise.")

    df["Norm_Num_Ratings"] = df["Num_Ratings"] / max_n
    df["Norm_Avg_Rating"]  = df["Avg_Rating"] / max_r
    df["Weighted_Score"]   = w1*df["Norm_Num_Ratings"] + w2*df["Norm_Avg_Rating"]
    return df


def add_bayesian_score(df, ratings_df):
    """
    Adds column 'Bayesian_Rating' to *df* using standard Bayesian averaging.
    """
    m = ratings_df["rating"].mean()                 # global mean
    C = df["Num_Ratings"].mean()                    # average vote-count

    df["Bayesian_Rating"] = (
        (C * m + df["Num_Ratings"] * df["Avg_Rating"])
        / (C + df["Num_Ratings"])
    )
    return df


# =============================================================
# 2. COMPUTE BOTH SCORES
# =============================================================
anime_stats = add_weighted_score(anime_stats, w1=0.5, w2=0.5)
anime_stats = add_bayesian_score(anime_stats, rating_df_clean)

# normalise both columns to 0-1 before blending
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
anime_stats[["Weighted_norm", "Bayesian_norm"]] = scaler.fit_transform(
    anime_stats[["Weighted_Score", "Bayesian_Rating"]]
)

# =============================================================
# 3. HYBRID  (70 % Bayesian, 30 % Weighted)
# =============================================================
alpha = 0.7          # give more weight to Bayesian
anime_stats["Hybrid_Score"] = (
    alpha   * anime_stats["Bayesian_norm"]
    + (1-alpha) * anime_stats["Weighted_norm"]
)

# -------------------------------------------------------------
TOP_N = 10
hybrid_recs = (
    anime_stats
      .sort_values("Hybrid_Score", ascending=False)
      .loc[:, ["anime_name", "Avg_Rating", "Num_Ratings",
               "Weighted_Score", "Bayesian_Rating", "Hybrid_Score"]]
      .head(TOP_N)
      .reset_index(drop=True)
)

print(f"\nTop {TOP_N} Anime â€“ Hybrid (70 % Bayesian, 30 % Weighted):\n")
print(hybrid_recs.to_string(index=False, formatters={
          "Avg_Rating":     "{:.2f}".format,
          "Weighted_Score": "{:.3f}".format,
          "Bayesian_Rating":"{:.3f}".format,
          "Hybrid_Score":   "{:.3f}".format,
      }))

# --- after computing `hybrid_recs` as above -------------------
top_names = hybrid_recs["anime_name"].tolist()

print(f"We recommend you these {TOP_N} Anime:")
for name in top_names:
    print(name)

# If you literally need the list object:
# return top_names   # inside a function
```
